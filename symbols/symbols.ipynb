```python
import schemdraw
import schemdraw.elements as elm
```

```python
# SMPS
schemdraw.theme('monokai')
with schemdraw.Drawing() as d:
    d.config(fontsize=12)
    d += elm.Dot().label("n1")
    d += elm.SourceV().label('V2: 0V').right().reverse()
    d.push()
    d += elm.SourceControlledI().down().label("B2")
    d += elm.Line().right()
    d.pop()
    d += elm.Line().label("n11")
    d += elm.SourceControlledI().down().label("B1")
    d += elm.Ground()
    d += elm.Gap().right()
    d += elm.Ground()
    d += elm.SourceV().label('V1: $V_{out}$', loc="bottom")
    d += elm.Line().right().label("n2")
    d += elm.Dot()
    d += elm.Gap().down()
    d += elm.Gap().right()
    d += elm.Ground()
    d += elm.SourceControlledV().label("BVO")
    d += elm.Dot().label("n3")
    d += elm.Gap().down()
    d += elm.Gap().right()
    d += elm.Ground()
    d += elm.SourceControlledV().label("BIO")
    d += elm.Dot().label("n4")
    d += elm.Gap().right()
    d += elm.Gap().down()
    d += elm.Ground()
    d += elm.SourceControlledV().label("BVI")
    d += elm.Dot().label("n5")
    d += elm.Gap().right()
    d += elm.Gap().down()
    d += elm.Ground()
    d += elm.SourceControlledV().label("BII")
    d += elm.Dot().label("n6")
    d += elm.Gap().right()
    d += elm.Gap().down()
    d += elm.Ground()
    d += elm.SourceControlledV().label("BEF")
    d += elm.Dot().label("n7")

d.save('SMPS.svg')
```
